```python
from selenium import webdriver
from selenium.webdriver.common.keys import Keys
import time

class tinderBot():
    def __init__(self):
        self.driver = webdriver.Chrome("C:\\Drivers\\chromedriver_win32\\chromedriver.exe")
        
    def login(self):
        self.driver.get('https://tinder.com/')
        
        time.sleep(5)
        fb_btn = self.driver.find_element_by_xpath('//*[@id="modal-manager"]/div/div/div/div/div[3]/div[2]/button')
        fb_btn.click()
        
        
        #As new pop up widow comes. This will handle the new window
        base_window = self.driver.window_handles[0]
        self.driver.switch_to.window(self.driver.window_handles[1])

        
        email_in = self.driver.find_element_by_id('email').send_keys('jyotishman.parasar888@gmail.com')
        pass_in = self.driver.find_element_by_id('pass').send_keys('kissmyassman')
        
        login_btn = self.driver.find_element_by_id('u_0_0')
        login_btn.click()
        
        self.driver.switch_to.window(base_window)
        
        time.sleep(2)
        
        popup_1 = self.driver.find_element_by_xpath('//*[@id="modal-manager"]/div/div/div/div/div[3]/button[1]')
        popup_1.click()
        
        time.sleep(3)

        popup_2 = self.driver.find_element_by_xpath('//*[@id="modal-manager"]/div/div/div/div/div[3]/button[1]')
        popup_2.click()
        
        
    def get_like(self):
        like_btn = self.driver.find_element_by_xpath('//*[@id="content"]/div/div[1]/div/main/div[1]/div/div/div[1]/div/div[2]/button[3]')
        like_btn.click()
        
        
    def dislike(self):
        dislike_btn = self.driver.find_element_by_xpath('//*[@id="content"]/div/div[1]/div/main/div[1]/div/div/div[1]/div/div[2]/button[1]')
        dislike_btn.click()
        
    def auto_swipe(self):
        while True:
            sleep(0.5)
            try:
                self.get_like()
            except Exception:
                try:
                    self.close_popup()
                except Exception:
                    self.close_match()

    def close_popup(self):
        popup_3 = self.driver.find_element_by_xpath('//*[@id="modal-manager"]/div/div/div[2]/button[2]')
        popup_3.click()
        
        popup_4 = self.driver.find_element_by_xpath('//*[@id="modal-manager"]/div/div/div[3]/button[2]')
        popup_4.click()
        
        

    def close_match(self):
        match_popup = self.driver.find_element_by_xpath('//*[@id="modal-manager-canvas"]/div/div/div[1]/div/div[3]/a')
        match_popup.click()
            
            
        
        
x = tinderBot()
x.login()
x.auto_swipe()
```
